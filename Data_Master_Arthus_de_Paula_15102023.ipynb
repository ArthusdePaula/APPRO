## Feature Selection

### Correlação

#### Correlação das Features entre si

```python
features_corr_matrix = train_features_df.corr()
```

```python
features_corr_matrix
```

```python
features_corr = features_corr_matrix.abs().unstack().sort_values(ascending=False).drop_duplicates().reset_index()
features_corr.rename(columns={0: "abs_correlation", "level_0": "feature_a", "level_1": "feature_b"}, inplace=True)
features_corr
```

```python
features_corr["is_redundant_98"] = features_corr["abs_correlation"] > 0.98
features_corr["is_redundant_95"] = features_corr["abs_correlation"] > 0.95
features_corr["is_redundant_90"] = features_corr["abs_correlation"] > 0.90
```

```python
features_corr["is_redundant_98"].sum(), features_corr["is_redundant_95"].sum(), features_corr["is_redundant_90"].sum()
```

#### Correlação das Features com a variável target

```python
features_to_target_correlation = pd.concat([train_features_df, target_df], axis=1).corr().iloc[:-1, -1]

features_to_target_correlation
```

```python
features_to_target_correlation = features_to_target_correlation.reset_index()
features_to_target_correlation
```

```python
features_to_target_correlation["irrelevant_01"] = features_to_target_correlation.TARGET.abs() < 0.01
features_to_target_correlation["irrelevant_001"] = features_to_target_correlation.TARGET.abs() < 0.001
features_to_target_correlation["irrelevant_0001"] = features_to_target_correlation.TARGET.abs() < 0.0001
```

```python
features_to_target_correlation
```

#### Comparando a lista de features com baixa correlação com o target e a lista de features com alta correlação entre si

```python
filtered_features_corr = features_corr.loc[features_corr["is_redundant_90"], ["feature_a", "feature_b"]]
```

```python
final_correlated = pd.merge(filtered_features_corr, features_to_target_correlation[["index", "TARGET"]], left_on="feature_a", right_on="index", how="left").drop(columns="index").rename({"TARGET": "feature_a_to_target"}, axis=1)
final_correlated = pd.merge(final_correlated, features_to_target_correlation[["index", "TARGET"]], left_on="feature_b", right_on="index", how="left").drop(columns="index").rename({"TARGET": "feature_b_to_target"}, axis=1)
final_correlated
```

```python
final_correlated[["feature_a_to_target", "feature_b_to_target"]] = final_correlated[["feature_a_to_target", "feature_b_to_target"]].abs()
final_correlated
```

```python
final_correlated["a_greater_corr_to_target"] = final_correlated["feature_a_to_target"] >= final_correlated["feature_b_to_target"]
final_correlated
```

```python
features_to_drop = final_correlated.loc[final_correlated.a_greater_corr_to_target, "feature_b"].values.tolist()
features_to_drop.extend(final_correlated.loc[~final_correlated.a_greater_corr_to_target, "feature_a"].values.tolist())
features_to_drop
```

```python
features_to_drop = list(set(features_to_drop))
```

# Iniciando modelagem

```python
from sklearn.pipeline import Pipeline
import xgboost as xgb
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, roc_auc_score, f1_score, precision_score, recall_score, confusion_matrix
from imblearn.over_sampling import SMOTE

def calculate_metrics(train_ds, target_ds, objective, oversample=False):
    all_metrics = []

    from sklearn.model_selection import StratifiedKFold
    kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)
    for train_ix, test_ix in kfold.split(train_ds, target_ds):
        model = Pipeline(
            [
                ("scaler", StandardScaler()),
                ("xgb", xgb.XGBClassifier(random_state=42, n_jobs=-1, objective=objective, eval_metric="pre")),
            ]
        )
        train_X, test_X = train_ds.iloc[train_ix], train_ds.iloc[test_ix]
        train_y, test_y = target_ds.iloc[train_ix], target_ds.iloc[test_ix]

        if oversample:
            smote = SMOTE(random_state=42)
            X_train_smote, y_train_smote = smote.fit_resample(train_ds, target_ds)
            model.fit(X_train_smote, y_train_smote)
        else:
            model.fit(train_X, train_y)

        pred = model.predict(test_X)
        all_metrics.append(
            {
                "accuracy": accuracy_score(test_y, pred),
                "roc_auc": roc_auc_score(test_y, model.predict_proba(test_X)[:, 1]),
                "f1": f1_score(test_y, pred),
                "precision": precision_score(test_y, pred),
                "recall": recall_score(test_y, pred),
                "confusion_matrix": confusion_matrix(test_y, pred).ravel(),
            }
    )

    return pd.DataFrame(all_metrics)
```

### Comparando as métricas e objectives do modelo

```python
for objective in ["binary:logistic", "binary:logitraw", "binary:hinge"]:
    full_ds_metrics = calculate_metrics(train_features_df, target_df, objective)
    print(objective)
    print("tn, fp, fn, tp", full_ds_metrics["confusion_matrix"].sum(axis=0))
    print(full_ds_metrics[["accuracy", "roc_auc", "f1", "precision", "recall"]].mean())
```

## Letra C - Dados de Treino

```python
small_train_dfc = pd.concat([small_train_df, target_df], axis=1)
```

```python
from sklearn.cluster import KMeans
```

```python
# Normalizar os dados
scaler = StandardScaler()
scaled_data = scaler.fit_transform(small_train_df)
```

```python
# Criar k clusters
kmeans = KMeans(n_clusters=10, random_state=0)
clusters = kmeans.fit_predict(scaled_data)
```

```python
#fizemos elbow kmeans porém demonstrou que o k ideal seria entre 2 e 4,
#como a letra C pede os 3 maiores, decidimos utilizar um número arbitrario de 10 para selecionar os 3 com maior lucro.
```

```python
# Adicionar os clusters ao dataframe
small_train_dfc['Cluster'] = clusters
```

```python
expected_revenue = 100
expected_cost = 10

def calculate_profit(row):
    if row['TARGET'] >= 0.5:
        return expected_revenue - expected_cost
    else:
        return 0

small_train_dfc['Expected_Profit'] = small_train_dfc.apply(calculate_profit, axis=1)

# Agora, vamos calcular o lucro médio por cluster
profit_per_cluster = small_train_dfc.groupby('Cluster')['Expected_Profit'].mean()
```

```python
#top clusters para dados de treino
top_clusters = profit_per_cluster.sort_values(ascending=False).head(3)
print(top_clusters)
```
